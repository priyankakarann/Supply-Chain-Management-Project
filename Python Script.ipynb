```python
%pip install pymysql

import numpy as np
import pymysql
```

```python
myconnection = pymysql.connect(host='localhost', user='root', passwd='TIGER', database='youtube')
```

```python
cur = myconnection.cursor()
```

```python
df = pd.read_excel('supply-chain-final.xlsx')
```

```python
df.head(2)
```

```python
# Remove the first column using iloc
df = df.iloc[:, 1:]
```

```python
df.head(2)
```

```python
# Check the number of columns in your DataFrame
num_cols = len(df.columns)

# Generate a string with the same number of placeholders as the number of columns
placeholders = ','.join(['%s'] * num_cols)
```

```python
# create a new table into the database and insert the data.
cur.execute("create table cars1 (SupplierID text, SupplierAddress text, SupplierName text, SupplierContactDetails text, ProductID text, CarMaker text, CarModel text, CarColor text, CarModelYear text, CarPrice text, CustomerID text, CustomerName text, Gender text, JobTitle text, PhoneNumber text, EmailAddress text, City text, Country text, CountryCode text, State text, CustomerAddress text, OrderDate text, OrderID text, ShipDate text, ShipMode text, Shipping text, PostalCode text, Sales text, Quantity text, Discount text, CreditCardType text, CreditCard text, CustomerFeedback text)")
```

```python
sql = "INSERT INTO cars VALUES ({})".format(placeholders)
```

```python
# convert every record/row into tuple and commit it.
    
for i in range(len(df)):
    cur.execute(sql, tuple(df.iloc[i]))
    myconnection.commit()
```
